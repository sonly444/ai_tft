```python
# MongoDB 접속 방법

# MongoDB 접속 패키지
from pymongo import MongoClient

# mongodb_URI = "mongodb://<IP>:<Port>/"
# client = MongoClient(mongodb_URI)

# MongoDB 접속 정보
client = MongoClient(
    host="10.1.10.145",
    port=27017,
    username="ipamuser",
    password="ipamuser",
    authSource="ipamdb"
)

# 접속 테스트: 데이터베이스 목록 출력
print(client.list_database_names())
```

```python
# DB 접근
db = client['ipamdb']

print(db)
```

```python
# Collection (Table) 접근

iplist = db.iplist

print(iplist)
```

```python
# Collection 리스트 조회

db.list_collection_names()
```

```python
# Collection 내 단건 Document 조회 (샘플조회?)
import pprint

iplist = db.iplist

pprint.pprint(iplist.find_one())
```

```python
# Collection 내 단건 Document 조회 (조건 조회)

iplist = db.iplist

pprint.pprint(iplist.find_one({"ipaddr": "10.1.10.133"}))
```

```python
# 다중 Documents 조회 - find() Method 사용

# Collection 내 모든 Documents 조회

iplist = db.iplist

for results in iplist.find():
    pprint.pprint(results)
```

```python
# 다중 Documents 조회 - find() Method 사용

iplist = db.iplist

for results in iplist.find({"vmware": "N"}):
    pprint.pprint(results)

```

```python
# Counting

# Collection 내 도큐먼트 수 조회
iplist.count_documents({})
```

```python
# 조건을건을 통한 도큐먼트 수 조회
iplist.count_documents({"vmware": "N"})
```

```python
#범위 조회

iplist = db.iplist

for results in iplist.find({"vmware": "N"}):
    pprint.pprint(results)
```

```python
#IP 범위 조회 테스트
# MongoDB는 String 형태는 단순 알파벳 순으로 정렬함

iplist = db.iplist

for results in iplist.find({"ipaddr": {"$gte": "10.1.10.91", "$lte": "10.1.10.99"}}):
    pprint.pprint(results)


'''
# 검색 오류 예시 (String으로 비교하기 때문에, 아래와 같이 자릿수가 달라지는 경우 검색 결과가 없음)
for results in iplist.find({"ipaddr": {"$gte": "10.1.10.91", "$lte": "10.1.10.100"}}):
    pprint.pprint(results)

※ 해결 방법 고민 : IP의 경우 DB에 저장할 때 부터, 4자리의 컬럼을 나눈 다음 정수형으로 저장 ?
'''
```
